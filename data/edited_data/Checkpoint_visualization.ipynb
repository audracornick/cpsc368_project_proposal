```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
wait_times_df = pd.read_csv("preprocessed_surgical_wait_times.csv")
spending_df = pd.read_csv("preprocessed_health_data.csv")

years = []
i = 2009
while i < 2023:
    years.append(i)
    i = i+1
print('List of years:', years)
```

```python
## get total sum based on year
def total(groupBy: str, fromCol: str, dataFrame):
    group = dataFrame.groupby(groupBy)[fromCol].sum()
    listName = []
    for y in years:
        listName.append(group[y])
    return listName
```

```python
# list of total spending per year in chronological order
spending = total('year', 'total_current_dollars', spending_df) 

# list of total completed procedures per year in chronological order
completed = total('Year', 'COMPLETED', wait_times_df)  

# list of total waiting procedures per year in chronological order
waiting = total('Year', 'WAITING', wait_times_df)  
```

```python
plt.bar(years, completed, label='Total Completed Procedures')
plt.bar(years, waiting, label='Total Waiting Procedures')

plt.xlabel('Years')
plt.ylabel('Number of Procedures')
plt.title('Total completed and waiting prodedures over time')
plt.legend(['Total Completed Procedures', 'Total Waiting Procedures'], loc="upper left")
plt.show()
```

```python
## Separate charts
plt.bar(years, completed, label='Total Completed Procedures')

plt.xlabel('Years')
plt.ylabel('Number of Procedures')
plt.title('Total Completed Prodedures Over Time')
plt.legend(['Total Completed Procedures'], loc="upper left")
plt.show()
```

```python
plt.bar(years, waiting, label='Total Waiting Procedures', color='orange')

plt.xlabel('Years')
plt.ylabel('Number of Procedures')
plt.title('Total Waiting Prodedures Over Time')
plt.legend(['Total Waiting Procedures'], loc="upper left")
plt.show()
```
